# Review Bot

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Restaurant_Reviews.tsv', delimiter = '\t', quoting = 3)
```

```python
dataset.head()
```

## Cleaning the texts

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, 1000):
  review = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])
  review = review.lower()
  review = review.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
print(corpus)
```

## Creating the Bag of Words model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, -1].values
```

## Applying K-Means Clustering

```python
review_cat_df = pd.DataFrame(X)
print(review_cat_df)
```

```python
from sklearn.cluster import KMeans

# Use KMeans() function to create k-means clusters with k=6 (n_clusters=6). 
kmeans_4 = KMeans(n_clusters=4, random_state=0).fit(review_cat_df)

# Develop cluster membership for k-means clusters, k=6.
memb_km4 = pd.Series(kmeans_4.labels_, index=review_cat_df.index)

# Display cluster memberships for 6 clusters.  
print('Cluster Membership for 4 Clusters Using k-Means Clustering')
print()
cluster_dict = {}
for key, item in memb_km4.groupby(memb_km4):
    cluster_dict[key] = list(item.index)
    #print(key, ':', list(item.index))
    #print(key, ': ', ', '.join(str(item.index)))
```

```python
for k in cluster_dict.keys():
    print('Cluster-'+str(k)+':')
    for i in cluster_dict[k]:
        print(i,dataset['Review'][i])
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

## Training the Naive Bayes model on the Training set

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

## Predicting the Test set results For Naive Bayes Model

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix For Naive Bayes Model

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Training the Neural Network model on the Training set

```python
from sklearn.neural_network import MLPClassifier

review_nn = MLPClassifier(hidden_layer_sizes=(2), activation='logistic', solver='lbfgs', max_iter=10000)

review_nn.fit(X_train,y_train)
```

## Predicting the Test set results For Neural Network Model

```python
review_nn_result_df = pd.DataFrame(
    {
        'Actual': y_test,
        'Predicted': review_nn.predict(X_test)
    }
)
print(review_nn_result_df)
```
